```python
import detector as dt
import numpy as np
```

```python
photo_for_check = 'ss04.jpg'
```

```python
descriptor = dt.get_face_descriptor(photo_for_check)
print(descriptor)
```

```python
base_descriptors = np.load('registers/ss.npy')
print(base_descriptors)
```

```python
distances = np.linalg.norm(descriptor - base_descriptors, axis=1)
print(distances)
```

```python
min_index = np.argmin(distances)
min_distance = distances[min_index]
print(min_distance)
```

```python
threshold = 0.5

if min_distance < threshold:
    print("É o Slvio Santos")
else:
    print("Não é o Slvio Santos")
```
